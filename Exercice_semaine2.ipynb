```python
# Install and load packages

import pandas as pd
from matplotlib.pyplot import subplots

!pip install openpyxl
import openpyxl
```

```python
### 1 and 2. Upload the Vahatra_defor_noNA database and print it

pas = pd.read_excel('Vahatra_defor_noNA.xlsx')
pas

```

```python
### 3. Print list of dataframe columns
pas.columns
```

```python
### 4. Rename treecover columns
pas.columns = pas.columns.str.replace("-01-01_treecover_ha", "", regex=False)
pas.columns
```

```python
### 5.Compute the average deforestation rate over the period 2000-2023 and assign it to a new column.
pas['deforest_2000_2023'] = (pas['treecover_area_2000']-pas['treecover_area_2023'])/pas['treecover_area_2000']*100
```

```python
### 6. Scatter plot of deforestation rate and inital forested area
fig, axis = subplots(figsize=(8,8))
axis.scatter(pas['treecover_area_2000'],pas['deforest_2000_2023'], marker='o')
axis.set_xlabel("Treecover area in 2000 (ha)")
axis.set_ylabel("Deforestation rate 2000-2023")
axis.set_title("Plot of deforestation rate vs intial forested area")
```

```python
### 7. Scatter plot of deforestation rate and date of creation
fig, axis = subplots(figsize=(8,8))
axis.scatter(pas['date_creation'],pas['deforest_2000_2023'], marker='o')
axis.set_xlabel("Year")
axis.set_ylabel("Deforestation rate 2000-2023")
axis.set_title("Plot of deforestation rate vs year of creation")
```

```python
### 8. Create the same scatter plot as in question 7 but restrict your analysis to protected areas belonging to IUCN categories I and II.

#Filter using cat_iucn
categories = (pas['cat_iucn'] == 'I') | (pas['cat_iucn'] == 'II')
pas_re = pas.loc[categories, :]
print(pas_re['cat_iucn'])

#Plot using new df
fig, axis = subplots(figsize=(8,8))
axis.scatter(pas_re['date_creation'],pas_re['deforest_2000_2023'], marker='o')
axis.set_xlabel("Year")
axis.set_ylabel("Deforestation rate 2000-2023")
axis.set_title("Plot of deforestation rate vs year of creation")
```
