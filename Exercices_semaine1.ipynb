## Exercice 1

En ajoutant, supprimant et modifiant des éléments, nettoyez la liste suivante pour qu’elle contienne les notes de musique « do ré mi fa sol la si » dans le bon ordre.

`l = ["solsi", "re", "re", "fa", "fa", "sol", "la", "solsi"]`

Indice : utilisez l’index de la liste, la fonction `del` et la fonction [append](https://mimo.org/glossary/python/append()).

Tapez votre réponse ci-dessous: 

```python
l = ["solsi", "re", "re", "fa", "fa", "sol", "la", "solsi"]
l[0] = "do"
l[3] = "mi"
del(l[2::5]) #Utilisation du slicer 
l.append("si")
print(l)
```

## Exercice 2

Étant donné le dictionnaire ci-dessous, affiche à l’aide d’instructions print :

1. La liste des noms d’équipes

2. Le score de Léa lors du match de football

3. La liste des performances de Max dans tous les matchs

4. La liste des joueurs de l’équipe « Tigres »

5. La liste des sports pratiqués par l’équipe « Lions »

6. La liste de tous les sports pratiqués par toutes les équipes

7. La liste des performances des joueuses dans les deux équipes

```python
equipes = {
    "Lions": {
        "Léa": {"performances": {"football": 3, "basketball": 7}},
        "Marc": {"performances": {"football": 1, "basketball": 5}}
    },
    "Tigres": {
        "Max": {"performances": {"football": 4, "tennis": 6}},
        "Julie": {"performances": {"football": 2, "tennis": 8}}
    }
}
```

```python
# 1
print("Noms d'équipes :", equipes.keys())

#2
print('Score de Léa lors du match de football :', equipes["Lions"]["Léa"]["performances"]["football"])

#3
print('Performances de Max dans tous les matchs :', list(equipes["Tigres"]["Max"]["performances"].values()))

#4
print('Joueurs de l’équipe « Tigres » :', list(equipes["Tigres"].keys()))

#5
print('Sports pratiqués par l’équipe « Lions » :', list(equipes["Lions"]["Léa"]["performances"].keys()))

#6
print('Sports pratiqués par toutes les équipes :', list(set(equipes["Lions"]["Léa"]["performances"].keys()) | set(equipes["Tigres"]["Max"]["performances"].keys())))

#6 version alternative
print("Sports pratiqués par toutes les équipes:", list(equipes["Lions"]["Léa"]["performances"] | equipes["Tigres"]["Max"]["performances"]))

#7
print('Performances des joueuses dans les deux équipes : Léa', equipes["Lions"]["Léa"]["performances"], 'Julie', equipes["Tigres"]["Julie"]["performances"])
```
